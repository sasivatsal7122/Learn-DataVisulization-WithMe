# DataVisualization

## PairPlot

Pairplot is a module of seaborn library which provides a high-level interface for drawing attractive and informative statistical graphics.Pairplot visualizes given data to find the relationship between them where the variables can be continuous or categorical.
Plot pairwise relationships in a data-set.Pairplot visualization comes handy when you want to go for Exploratory data analysis (“EDA”).


```python
pairplot = sns.pairplot(data=penguins_df,hue='species')
pairplot.fig.suptitle("Feature comparison and variable relation in Penguins dataset")
plt.tight_layout()
pairplot.fig.subplots_adjust(top=0.9)
```

## Joint Plot

 A Jointplot comprises three plots. Out of the three, one plot displays a bivariate graph which shows how the dependent variable(Y) varies with the independent variable(X). Another plot is placed horizontally at the top of the bivariate graph and it shows the distribution of the independent variable(X). The third plot is placed on the right margin of the bivariate graph with the orientation set to vertical and it shows the distribution of the dependent variable(Y). It is very helpful to have univariate and bivariate plots together in one figure.

```python
penguins_df.dropna(inplace=True)
penguins_df['species-gen'] = penguins_df.apply(lambda x : '{}-{}'.format(x['species'],x['sex']),axis=1)
penguins_df
```

```python
sns.jointplot(data=penguins_df, x='bill_length_mm',y='body_mass_g',hue='species-gen',kind='scatter',palette='Paired',height=10)
```

```python
sns.jointplot(data=penguins_df, x='bill_length_mm',y='body_mass_g',hue='species-gen',kind='kde',palette='Paired',height=10)
```

```python
sns.jointplot(data=penguins_df, x='bill_length_mm',y='body_mass_g',hue='species-gen',kind='kde',shade=True,alpha=0.5,palette='husl',height=10)
```

```python
jointgrid = sns.jointplot(data=penguins_df,x='bill_length_mm',y='bill_depth_mm',hue='species-gen',kind='scatter',palette='magma',height=10)
jointgrid.plot_joint(sns.kdeplot,shade=True,alpha=0.4)
```
